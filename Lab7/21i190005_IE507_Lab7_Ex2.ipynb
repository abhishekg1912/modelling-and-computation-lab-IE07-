#Que.3

```python
plots=sns.barplot(x='mpg',y='displacement',data=df)
x=plots.patches[3].get_x() + plots.patches[3].get_width() / 2.
y=plots.lines[3].get_ydata()[1]
plt.annotate('Highest bar in the plot.', xy=(x,y), xytext=(x+0.3,y+55.5), arrowprops=dict(facecolor='green'))
```

In this bar plot we can observe that the mpg is inversely proportional to the displacement i.e. the displacement of cars decrease with increasing of the mpg.

#Que.4

```python
print(f"The maximum value of the displacement {max(df.displacement)}")
print(len(df.mpg))
```

```python
num_of_maximum_bar=0
value_of_mpg=[]
for i in range(398):
  if df.displacement[i]==455:
    num_of_maximum_bar+=1
    value_of_mpg.append(i)
  else:
    pass
print(f"number of lower bars = {num_of_maximum_bar}")
print(f"list of index of mpg for lower bars ={value_of_mpg}")
for i in value_of_mpg:
  print(f"The value of mpg= {df.mpg[i]}")


```

```python
ax=sns.barplot(x='mpg',y='displacement',data=df,hue='model_year',estimator=np.median)
ax.legend(loc='upper right',ncol=4)
ax.set_xlabel("mpg",fontsize=15)
ax.set_ylabel("Displacement",fontsize=15)
ax.annotate(np.median(df.displacement),xy=(14, 455),xytext = (14, 455))
```

```python
print(f"The minimum value of the displacement {min(df.displacement)}")
print(len(df.mpg))
```

```python
num_of_lower_bar=0
value_of_mpg=[]
for i in range(398):
  if df.displacement[i]==68:
    num_of_lower_bar+=1
    value_of_mpg.append(i)
  else:
    pass
print(f"number of lower bars = {num_of_lower_bar}")
print(f"list of index of mpg for lower bars ={value_of_mpg}")
print(df.mpg[117])
```

```python
ax=sns.barplot(x='mpg',y='displacement',data=df,hue='origin',estimator=np.median)
ax.legend(loc='upper right')
ax.set_xlabel("mpg",fontsize=15)
ax.set_ylabel("Displacement",fontsize=15)
ax.annotate(np.median(df.displacement),xy=(df.mpg[117], 68),xytext = (df.mpg[117], 68))
```

The above curve shows that the cars with origin Usa have higher displacement with lower mpg.

#Que.10

```python
sns.boxplot(x='model_year',y="displacement",hue='origin',data=df)
plt.grid()
```

Cars with origin USA have much greater displacement than cars with origin Japan and Europe.The median displacement for Usa originated cars remain same in model year 74 to 77 and after model year 77 the median displacement starts decreasing.
