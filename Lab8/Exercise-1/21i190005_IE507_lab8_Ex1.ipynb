#Que.7

To find the best choice of clusters we will use the elbow method in sense of distortion of data points, in this method it calculates the avg. distortion of data points for each no. of clusters and we will increase the number of clusters untill we get effective decrease in distortion if rate of decrement of distortion becomes low then we will not increase the no. of clusters and that no. of clusters becomes the best choice of the no of clusters. 

#Que.8

```python
distortions = []
for k in range(3,20):
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)

plt.plot(range(3,20), distortions, 'ro-')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
#plt.title('The Elbow Method showing the optimal number of clusters for data of file s1.txt');
```

In the above graph we can see that the rate of decrement of the graph of distortion becomes low at 8 clusters, so 8 clusters will be the best choice of no. of clusters.

#Que.19

Normalization makes sure that all of your data looks and reads the same way across all records.normalising the features around the center 0 and with a standard deviation of 1 is important when we compare measurements that have different units. Variables that are measured at different scales do not contribute equally to the analysis and might end up creating a bais.

Example:
A variable that ranges between 0 and 1000 will outweigh a variable that ranges between 0 and 1. Using these variables without normalization will give the variable with the larger range weight of 1000 in the analysis. Transforming the data to comparable scales can prevent this problem. Typical data normalization procedures equalize the range and/or data variability.
